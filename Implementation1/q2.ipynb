```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.special import gamma
```

```python
x_array = np.random.uniform(low=0 , high=1 , size = 10000)
```

```python
print(len([i for i in x_array if i>0.5]))
```

```python
def beta_prob(alpha,beta,theta):
    beta_alpha_beta = gamma(alpha+beta)/ ( gamma(alpha)*gamma(beta) )
    p1 = theta**(alpha-1)
    p2 = (1-theta)**(beta-1)
    return (p1*p2) * (beta_alpha_beta)
```

```python
all_beta_values = list(map(lambda i:beta_prob(4,5,i),x_array))
```

```python
print(all_beta_values[1:2])
```

```python
%matplotlib.inline
```

```python
plt.figure(figsize=(15,6))
plt.scatter(y=all_beta_values,x=x_array)
```

```python
new_beta_values = list(map(lambda i:beta_prob(22,32,i),x_array))
```

```python
plt.figure(figsize=(15,6))
plt.scatter(y=new_beta_values,x=x_array)
```

```python

val = list(map(lambda i:beta_prob(42,62,i),x_array))
```

```python
plt.figure(figsize=(15,6))

plt.scatter(y=val,x=x_array)
```

As the number of coin tosses increase the peak seems to  become more sharp around the true probability i.e 0.4.
