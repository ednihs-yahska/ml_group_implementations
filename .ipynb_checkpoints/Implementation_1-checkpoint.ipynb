```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import random
%matplotlib inline

dev_data = pd.read_csv("PA1_dev.csv.xls")

def dateSplitter(d, i):
    dateElements = d.split('/')
    return int(dateElements[i])

def stats(dataframe):
    stats = dataframe.describe()
    stats = stats.drop(["count", "25%", "50%", "75%",], 0)
    stats.loc["range"] = stats.loc["max"] - stats.loc["min"]
    return stats

def ordinal_stats(dataframe, args):
    stats = dataframe.describe()
    frequency = [pd.Series(dataframe[x].value_counts() / dataframe[x].count()) for x in args]
    return pd.DataFrame(frequency)
    
    

plt.rcParams["figure.figsize"] = [20,10]
price = dev_data["price"]
dev_data = dev_data[["id","date","bedrooms","bathrooms", "sqft_living", "sqft_lot", "floors", "waterfront", "view", "condition","sqft_above","sqft_basement","yr_built","yr_renovated","zipcode", "grade", "sqft_living15","sqft_lot15","price"]]
dev_data_continous = dev_data[["id","bedrooms","bathrooms", "sqft_living", "sqft_lot", "floors", "view", "condition","sqft_above","sqft_basement","yr_built","yr_renovated","zipcode", "sqft_living15","sqft_lot15","price"]]
dev_data_ordinal = dev_data[["waterfront", "grade", "condition"]]
dev_data.head()
print(dev_data.shape)



```

```python
dev_data_normalized = (dev_data - dev_data.min()) / (dev_data.max() - dev_data.min())
dev_data_normalized
```

```python
W = [random.randint(0,99) for x in range(0, len(dev_data_normalized.columns))]
print(len(W))
```

```python
df = dev_data_normalized
Y_ = []
X = []
norm = np.sum(W)
epsilon = 0.5
price = (price - price.min()) / (price.max() - price.min())
Yi = np.array(price)
iteration = 1
while(norm > epsilon):
    gradient = np.array(np.zeros(df.shape[1]))
    for i in range(0, len(df.index)):
        Xi = np.array(df.iloc[i])
        X.append(Xi)
        iy_ = np.array(np.transpose(W)).dot(np.array(Xi))
        yi = Yi[i]
        WtX_Y = iy_ - yi
        y_ = WtX_Y*Xi
        gradient = gradient + y_
    W = W - 0.0001*gradient
    norm = math.sqrt(np.transpose(gradient).dot(gradient))
    Y_=[]
    X=[]
    iteration+=1
print(W)

    

    

    


    
    
```
