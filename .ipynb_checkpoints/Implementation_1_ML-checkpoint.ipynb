```python
#importing required packages and loading training data.
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
training_data_source = 'PA1_train.csv.xls'
main_df = pd.read_csv(training_data_source)
plt.rcParams["figure.figsize"] = [20,10]
```

# Part 0

# correlation matrix

```python
#finding the correlation matrix
corr = main_df.corr()
corr.style.background_gradient().set_precision(2)
```

# 0 a:

Id feature is a **identifier** for each record and dependent on the sequence in which the datapoints are collected and as seen in above [correlation matrix](#correlation-matrix), **the correlation between id and price is very low**. Hence it is a bad idea to include id feature in learning.

```python
#dropping id feature
df = main_df.drop("id", 1)
df.sample(5)
```

```python
#Splitting date in month, day and year
df["date"] = pd.to_datetime(df["date"])
df["month_of_year"] = df['date'].apply(lambda x: x.month)
df["day_of_month"] = df['date'].apply(lambda x: x.day)
df["year"] = df['date'].apply(lambda x: x.year)
df.sample(5)
```

# 0 b:

We can use the month feature to determine if different seasons have any effect on the prices of the house.

# 0 c:

```python
#helper functions
def stats(dataframe):
    stats = dataframe.describe()
    stats = stats.drop(["count", "25%", "50%", "75%",], 0)
    stats.loc["range"] = stats.loc["max"] - stats.loc["min"]
    return stats

def ordinal_stats(dataframe):
    args = dataframe.columns.tolist()
    stats = dataframe.describe()
    frequency = [pd.Series((dataframe[x].value_counts() / dataframe[x].count())*100) for x in args]
    f = pd.DataFrame(frequency)
    f = f.fillna(0)
    return f

numerical_df = df.drop(["waterfront", "grade", "condition"], 1)
stats(numerical_df)
```

```python
df.sample(5)
```

```python
non_numerical_df = df[["waterfront", "grade", "condition"]]
ordinal_stats(non_numerical_df)
```

```python
df.sample(5)
```

# 0 e:
**Normalizing dataset**
<br/>
`Here we need to drop dummy before normalizing as it contains 1 in all the values which causes divide by zero error. We can add dummy back again after normalization.`

```python
#dropping price and date
df = df.drop(["price", "date"],1)
#dropping dummy before normalization and adding it back later
df = df.drop(["dummy"],1)
```

```python
df = (df - df.min()) / (df.max() - df.min())
```

```python
#adding back dummy feature
df["dummy"] = main_df["dummy"]
df.sample(5)
```
